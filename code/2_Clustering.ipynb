```python
from IPython.display import display_html
import pandas as pd

# a function to display multiple dataframes side by side
def display_side_by_side(*args, titles=()):
    html_str = ''
    for i, df in enumerate(args):
        title = titles[i] if i < len(titles) else f'DF{i+1}'
        html_str += f'<div style="display:inline-block; margin-right:20px;">'
        html_str += f'<h3>{title}</h3>'
        html_str += df.to_html()
        html_str += '</div>'
    display_html(html_str, raw=True)

df = pd.read_csv('../data/df_merged.csv', index_col=0)
display_side_by_side(
    df.head().T, df.tail().T, titles=['Data (Head)', 'Data (Tail)']
)
```

### Clustering

```python
# -*- coding: utf-8 -*-
'''
Clustering pipeline (K-Means & Hierarchical with cosine distance) - NO PCA
- Data prep (numeric-only, standardization)
- Model fit (multiple k), silhouette evaluation
- Visualizations (bars, 2D scatter on selected features, dendrogram, silhouette plot)
- Regime cross-tabs (USREC/ZLB/COVID vs clusters)
- Cluster-level feature means (plus heatmap) for interpretation
- FIXED COLORS: consistent colors across scatter and silhouette by cluster id
'''

# =========================
# 0) Imports
# =========================
import os
os.environ['OMP_NUM_THREADS'] = '1'
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples

from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# =========================
# 1) Data prep (numeric-only, impute, standardize)
# =========================
regime_cols = [c for c in ['USREC', 'ZLB_dummy', 'COVID_dummy'] if c in df.columns]

# Feature candidates based on macro theme (inflation, labor, real activity, rates, liquidity, etc.)
feature_cols_all = [
    'PCEPILFE_YoY','PCEPILFE_MoM','UNRATE','NROU',
    'GDPC1','GDPPOT','INDPRO','RSAFS',
    'MCOILWTICO','IR','OPHNFB','CES0500000003','CIVPART','JTSJOL',
    'FEDFUNDS','M2SL','WALCL','T5YIE','T10YIE','PX_MD','PX5_MD'
]
feature_cols = [c for c in feature_cols_all if c in df.columns]

X_df = df[feature_cols].copy()

scaler = StandardScaler()
X = scaler.fit_transform(X_df.values)

# Also keep a standardized DataFrame for plotting convenience
X_std = pd.DataFrame(X, columns=feature_cols)

# =========================
# 2) K-Means for multiple k and silhouette scores (Euclidean)
# =========================
k_list = list(range(2, 20))
kmeans_models = {}
kmeans_sil = {}

for k in k_list:
    km = KMeans(n_clusters=k, random_state=42, n_init=20)
    labels = km.fit_predict(X)
    if len(np.unique(labels)) > 1:
        kmeans_models[k] = (km, labels)
        kmeans_sil[k] = silhouette_score(X, labels, metric='euclidean')
    else:
        kmeans_sil[k] = np.nan

best_k_kmeans = max(kmeans_sil, key=lambda kk: (-1 if kmeans_sil[kk] is None else kmeans_sil[kk]))
km, labels_km = kmeans_models[best_k_kmeans]

# =========================
# 3) Hierarchical clustering (cosine distance) and silhouette by k
# =========================
D_cos = pdist(X, metric='cosine')           # distance = 1 - cosine similarity
Z = linkage(D_cos, method='average')        # average linkage on cosine distances

hclust_sil = {}
hclust_labels_dict = {}
for k in k_list:
    labels_h = fcluster(Z, t=k, criterion='maxclust')
    labels_h = labels_h - 1  # make labels start at 0
    if len(np.unique(labels_h)) > 1:
        hclust_labels_dict[k] = labels_h
        hclust_sil[k] = silhouette_score(X, labels_h, metric='cosine')
    else:
        hclust_sil[k] = np.nan

best_k_hclust = max(hclust_sil, key=lambda kk: (-1 if hclust_sil[kk] is None else hclust_sil[kk]))
labels_h_best = hclust_labels_dict[best_k_hclust]

# =========================
# 4) Visualizations
# =========================

# Make Save Folder
save_dir = '../results/2_Clustering'
os.makedirs(save_dir, exist_ok=True)

colors = list(range(20))

# ---- Fixed color helpers ----
def get_cluster_colors(labels, cmap=cm.tab20):
    """Return a dict {cluster_id: color} with a fixed color per cluster id."""
    uniq = np.unique(labels)
    colors = cmap(np.linspace(0, 1, len(uniq)))
    return {lab: col for lab, col in zip(uniq, colors)}

def plot_cluster_scatter_2d_fixed(X_df_std, x_col, y_col, labels, title, save_path, cluster_colors):
    """2D scatter on selected standardized features with fixed cluster colors."""
    plt.figure()
    plt.title(title)
    uniq = np.unique(labels)
    for lab in uniq:
        idx = labels == lab
        plt.scatter(X_df_std.loc[idx, x_col],
                    X_df_std.loc[idx, y_col],
                    s=18, label=f'Cluster {lab}', color=cluster_colors[lab])
    plt.xlabel(f'{x_col} (std)'); plt.ylabel(f'{y_col} (std)')
    plt.legend(loc='upper right', markerscale=1.5, frameon=True)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()

def plot_silhouette_fixed(X_clust, labels, title, save_path, cluster_colors, metric='euclidean'):
    """Silhouette plot with fixed colors per cluster id."""
    sample_sil_vals = silhouette_samples(X_clust, labels, metric=metric)
    plt.figure(figsize=(7, 5))
    plt.title(title)
    y_lower = 10
    uniq = np.unique(labels)
    for lab in uniq:
        ith = np.sort(sample_sil_vals[labels == lab])
        size_i = ith.shape[0]
        y_upper = y_lower + size_i
        plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith,
                          facecolor=cluster_colors[lab], edgecolor=cluster_colors[lab], alpha=0.7)
        plt.text(-0.05, y_lower + 0.5 * size_i, str(lab))
        y_lower = y_upper + 10
    plt.xlabel('Silhouette coefficient values'); plt.ylabel('Cluster label')
    plt.axvline(x=np.mean(sample_sil_vals), linestyle='--')
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()

# (a) KMeans silhouette scores (bar)
plt.figure()
plt.title('K-Means Silhouette Scores (Euclidean)')
plt.bar([str(k) for k in k_list], [kmeans_sil[k] for k in k_list])
plt.xlabel('k'); plt.ylabel('Silhouette score')
plt.tight_layout()
save_path = os.path.join(save_dir, 'kmeans_silhouette_scores.png')
plt.savefig(save_path, dpi=300)
plt.show()

# (b) Hierarchical (cosine) silhouette scores (bar)
plt.figure()
plt.title('Hierarchical (Cosine) Silhouette Scores')
plt.bar([str(k) for k in k_list], [hclust_sil[k] for k in k_list])
plt.xlabel('k'); plt.ylabel('Silhouette score')
plt.tight_layout()
save_path = os.path.join(save_dir, 'hierarchical_silhouette_scores.png')
plt.savefig(save_path, dpi=300)
plt.show()

# (c) 2D scatter on selected original features (standardized), color by KMeans best-k
viz_x = 'UNRATE' if 'UNRATE' in X_std.columns else X_std.columns[0]
viz_y = 'PCEPILFE_YoY' if 'PCEPILFE_YoY' in X_std.columns else X_std.columns[1]

kmeans_colors = get_cluster_colors(colors, cmap=cm.tab20)
save_path = os.path.join(save_dir, f'kmeans_clusters_k{best_k_kmeans}.png')
plot_cluster_scatter_2d_fixed(
    X_df_std=X_std,
    x_col=viz_x,
    y_col=viz_y,
    labels=labels_km,
    title=f'K-Means Clusters (k={best_k_kmeans}) on 2D features (standardized)',
    save_path=save_path,
    cluster_colors=kmeans_colors
)

# (d) Dendrogram (hierarchical clustering)
plt.figure(figsize=(10, 5))
plt.title('Hierarchical Clustering Dendrogram (Cosine, average)')
dendrogram(Z, no_labels=True, color_threshold=None)
plt.xlabel('Samples'); plt.ylabel('Distance')
plt.tight_layout()
save_path = os.path.join(save_dir, 'hierarchical_clustering_dendrogram.png')
plt.savefig(save_path, dpi=300)
plt.show()

# (e) 2D scatter on selected features, color by Hierarchical best-k
hclust_colors = get_cluster_colors(colors, cmap=cm.tab20)
save_path = os.path.join(save_dir, f'hierarchical_clusters_k{best_k_hclust}.png')
plot_cluster_scatter_2d_fixed(
    X_df_std=X_std,
    x_col=viz_x,
    y_col=viz_y,
    labels=labels_h_best,
    title=f'Hierarchical Clusters (k={best_k_hclust}) on 2D features (standardized)',
    save_path=save_path,
    cluster_colors=hclust_colors
)

# =========================
# 5) Save cluster assignments
# =========================
out_df = df.copy()
out_df['kmeans_k'] = best_k_kmeans
out_df['kmeans_label'] = labels_km
out_df['hclust_k'] = best_k_hclust
out_df['hclust_label'] = labels_h_best

out_path = os.path.join(save_dir, 'cluster_assignments.csv')
try:
    out_df.to_csv(out_path, index=True)
    print(f'[INFO] Cluster assignments saved to: {out_path}')
except Exception as e:
    print(f'[WARN] Failed to save cluster assignments: {e}')

# =========================
# 6) Save summary tables
# =========================
try:
    kmeans_means.to_csv(os.path.join(save_dir, 'kmeans_cluster_feature_means.csv'))
    hclust_means.to_csv(os.path.join(save_dir, 'hclust_cluster_feature_means.csv'))
    print('\n[INFO] Saved cluster feature means CSVs to ' + save_dir)
except Exception as e:
    print(f'[WARN] Failed to save feature means: {e}')
```

```python
# =========================
# 9) Per-regime clustering 2D plots (split by binary variable value)
# =========================

def cluster_and_plot_by_binary(df_all, regime_col, base_save_dir, 
                               k_max, random_state=42):
    """Split by a binary regime variable (0/1), run clustering within each subset,
    and save 2D scatter plots for both K-Means and Hierarchical."""
    assert regime_col in df_all.columns, f"{regime_col} not in dataframe."

    # Create parent folder for this regime variable
    parent_dir = os.path.join(base_save_dir, f'by_regime_{regime_col}')
    os.makedirs(parent_dir, exist_ok=True)

    # Iterate over regime values (e.g., 0 and 1)
    for val in sorted(df_all[regime_col].dropna().unique().tolist()):
        sub = df_all[df_all[regime_col] == val].copy()
        n = sub.shape[0]
        if n < 10:
            print(f"[WARN] Skip {regime_col}={val}: too few samples (n={n})")
            continue

        # ----- Data prep in subset (avoid leakage) -----
        X_sub_df = sub[feature_cols].copy()
        X_sub_df = X_sub_df.fillna(X_sub_df.mean(numeric_only=True))
        scaler_sub = StandardScaler()
        X_sub = scaler_sub.fit_transform(X_sub_df.values)
        X_sub_std = pd.DataFrame(X_sub, columns=feature_cols, index=X_sub_df.index)

        # Choose axes for plotting
        viz_x = 'UNRATE' if 'UNRATE' in X_sub_std.columns else X_sub_std.columns[0]
        viz_y = 'PCEPILFE_YoY' if 'PCEPILFE_YoY' in X_sub_std.columns else X_sub_std.columns[1]

        # Reasonable k range for subset size
        k_candidates = list(range(2, max(3, min(k_max, n - 1)) + 1))

        # ----- KMeans: pick best k by silhouette (Euclidean) -----
        km_models, km_sil = {}, {}
        for k in k_candidates:
            km = KMeans(n_clusters=k, random_state=random_state, n_init=20)
            lab = km.fit_predict(X_sub)
            if len(np.unique(lab)) > 1:
                km_models[k] = (km, lab)
                km_sil[k] = silhouette_score(X_sub, lab, metric='euclidean')
            else:
                km_sil[k] = np.nan
        best_k_km = max((k for k in k_candidates if k in km_models),
                        key=lambda kk: (km_sil.get(kk, -np.inf) 
                                        if not np.isnan(km_sil.get(kk, np.nan)) else -np.inf))
        _, labels_km_sub = km_models[best_k_km]
        km_colors_sub = get_cluster_colors(labels_km_sub, cmap=cm.tab20)

        # Save KMeans 2D scatter
        sub_dir = os.path.join(parent_dir, f'{regime_col}_{val}')
        os.makedirs(sub_dir, exist_ok=True)
        plot_cluster_scatter_2d_fixed(
            X_df_std=X_sub_std,
            x_col=viz_x,
            y_col=viz_y,
            labels=labels_km_sub,
            title=f'K-Means (k={best_k_km}) on {viz_x} vs {viz_y} | {regime_col}={val}',
            save_path=os.path.join(sub_dir, f'kmeans_scatter.png'),
            cluster_colors=km_colors_sub
        )

        # ----- Hierarchical (cosine): pick best k by silhouette -----
        D_cos_sub = pdist(X_sub, metric='cosine')
        Z_sub = linkage(D_cos_sub, method='average')
        hc_labels_dict, hc_sil = {}, {}
        for k in k_candidates:
            lab_h = fcluster(Z_sub, t=k, criterion='maxclust') - 1
            if len(np.unique(lab_h)) > 1:
                hc_labels_dict[k] = lab_h
                hc_sil[k] = silhouette_score(X_sub, lab_h, metric='cosine')
            else:
                hc_sil[k] = np.nan
        best_k_hc = max((k for k in k_candidates if k in hc_labels_dict),
                        key=lambda kk: (hc_sil.get(kk, -np.inf)
                                        if not np.isnan(hc_sil.get(kk, np.nan)) else -np.inf))
        labels_hc_sub = hc_labels_dict[best_k_hc]
        hc_colors_sub = get_cluster_colors(labels_hc_sub, cmap=cm.tab20)

        # Save Hierarchical 2D scatter
        plot_cluster_scatter_2d_fixed(
            X_df_std=X_sub_std,
            x_col=viz_x,
            y_col=viz_y,
            labels=labels_hc_sub,
            title=f'Hierarchical (k={best_k_hc}) on {viz_x} vs {viz_y} | {regime_col}={val}',
            save_path=os.path.join(sub_dir, f'hclust_scatter.png'),
            cluster_colors=hc_colors_sub
        )

        # Optional: save subset assignments
        sub_out = sub.copy()
        sub_out['kmeans_k_sub'] = best_k_km
        sub_out['kmeans_label_sub'] = labels_km_sub
        sub_out['hclust_k_sub'] = best_k_hc
        sub_out['hclust_label_sub'] = labels_hc_sub
        sub_out.to_csv(os.path.join(sub_dir, 'cluster_assignments_subset.csv'))

        print(f"[INFO] {regime_col}={val}: best K-Means k={best_k_km} (sil={km_sil[best_k_km]:.3f}), "
              f"best HClust k={best_k_hc} (sil={hc_sil[best_k_hc]:.3f}) -> plots saved in {sub_dir}")

# ---- Run for each binary regime variable present ----
for rc in ['USREC', 'ZLB_dummy', 'COVID_dummy']:
    if rc in out_df.columns:
        cluster_and_plot_by_binary(out_df, rc, save_dir, k_max=20)
    else:
        print(f"[WARN] {rc} not found; skip.")
```

```python
def plot_binary_pies_by_cluster(df, label_col, binary_cols, save_dir, fname_prefix, cmap=cm.tab20):
    """
    For each cluster (label_col), plot donut charts of 0/1 shares for all binary_cols.
    - Uses fixed cluster color (tab20) for "1" and gray for "0"
    - Ensures a visible boundary even at 0%/100% via a tiny epsilon slice
    - Adds a center annotation "0: n0 | 1: n1" for unambiguous reading
    """
    clusters = sorted(df[label_col].dropna().unique(), key=lambda x: (isinstance(x, str), x))
    n_clusters = len(clusters)
    n_cols = len(binary_cols)

    fig_w = max(3.2 * n_cols, 8)
    fig_h = max(2.8 * n_clusters, 6)
    fig, axes = plt.subplots(n_clusters, n_cols, figsize=(fig_w, fig_h), squeeze=False)

    cluster_colors = get_cluster_colors(clusters, cmap=cmap)

    for r, cl in enumerate(clusters):
        sub = df[df[label_col] == cl]
        for c, col in enumerate(binary_cols):
            ax = axes[r, c]

            # Coerce boolean to int, drop NaNs
            col_vals = sub[col].dropna()
            if col_vals.dtype == bool:
                col_vals = col_vals.astype(int)

            # Original counts (for clear center annotation)
            orig_counts = col_vals.value_counts().reindex([0, 1], fill_value=0).astype(float)
            total = orig_counts.sum()
            if total == 0:
                ax.axis('off')
                ax.set_title(f'{col} (no data)')
                continue

            # Add epsilon so 0%/100% still draws a boundary
            counts = orig_counts.copy()
            epsilon = 1e-6
            if counts[0] == 0:
                counts[0] = epsilon
                counts[1] = total - epsilon
            elif counts[1] == 0:
                counts[1] = epsilon
                counts[0] = total - epsilon

            vals = counts.values

            # Colors: gray for 0, cluster color for 1
            slice_colors = ['#dddddd', cluster_colors[cl]]

            # Donut style + white edge to emphasize the boundary
            wedges, texts, autotexts = ax.pie(
                vals,
                labels=None,                      # keep the plot clean; we add a shared legend
                autopct=lambda p: f'{p:.1f}%',
                startangle=90,
                counterclock=False,
                colors=slice_colors,
                explode=(0.02, 0.02),
                textprops={'fontsize': 9},
                wedgeprops={'width': 0.35, 'edgecolor': 'white', 'linewidth': 1.0}
            )

            # Center annotation: show exact counts to disambiguate 0% vs 100%
            n0, n1 = int(orig_counts[0]), int(orig_counts[1])
            ax.text(
                0, 0,
                f'0: {n0}\n1: {n1}',
                ha='center', va='center', fontsize=10, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.25', fc='white', ec='none', alpha=0.75)
            )

            ax.set_title(f'{col} — Cluster {cl}', fontsize=11)
            ax.axis('equal')
    
    if label_col=='kmeans_label':
        graph_title = 'K-Means Clustering'
    elif label_col=='hclust_label':
        graph_title = 'Hierarchical Clustering'
    else:
        raise ValueError(f"Unknown label_col: {label_col}. Must be 'kmeans_label' or 'hclust_label'.")
    fig.suptitle(f'{graph_title}: within-cluster shares of 0/1', fontsize=14, y=1.02)

    # Shared legend: explicit mapping for 0/1
    handles = [
        plt.Line2D([0], [0], marker='o', color='w', label='0', markerfacecolor='#dddddd', markersize=10),
        plt.Line2D([0], [0], marker='o', color='w', label='1 (cluster color)', markerfacecolor='black', markersize=10),
    ]
    fig.legend(handles=handles, loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.02))

    plt.tight_layout()
    os.makedirs(save_dir, exist_ok=True)
    out_path = os.path.join(save_dir, f'{fname_prefix}_{label_col}_binary_pies.png')
    plt.savefig(out_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f'Saved: {out_path}')

plot_binary_pies_by_cluster(out_df, 'kmeans_label', ['USREC','ZLB_dummy','COVID_dummy'], save_dir, 'pies')
plot_binary_pies_by_cluster(out_df, 'hclust_label', ['USREC','ZLB_dummy','COVID_dummy'], save_dir, 'pies')
```
